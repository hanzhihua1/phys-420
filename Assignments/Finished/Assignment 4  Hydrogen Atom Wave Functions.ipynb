# Hydrogen Atom Radial Wavefunctions
## Ordinary differential eigenvalue problem


Eigenvalue problems are boundary value problems that only have solutions for particular values of a parameter.  In quantum mechanics these parameters are the quantized variables- like the energy or angular momentum.  

The hydrogen atom is a standard problem in quantum mechanics. You
basically solve the Schroedinger equation for a 1/r central potential.
We won't go through the separation of variables, but you can separate
variables and end up with a differential equation 
$$
\left(\frac{d^{2}}{dr^{2}}+\frac{2}{r}\frac{d}{dr}\right)R(r)+\frac{2\mu}{\hbar^{2}}\left(E+\frac{Ze^{2}}{r}-\frac{l(l+1)\hbar^{2}}{2\mu r^{2}}\right)R(r)=0
$$
where $R$ is the radial wavefunction, $r$ the radius, $\mu$ the reduced
mass, $Ze$ the charge of the nucleus, $e$ the charge of the electron
and $l$ the angular momentum quantum number.

The first step is to define a new independent variable $\rho$ and
a constant $\lambda$ :
$$
\begin{array}{cc}
\rho=\sqrt{\frac{-8\mu E}{\hbar^{2}}}r & \lambda=\frac{Ze^{2}}{\hbar}\sqrt{\frac{-\mu}{2E}}\end{array};
$$ 

so the equation becomes:

$$
\frac{d^{2}R}{d\rho^{2}}+\frac{2}{\rho}\frac{dR}{d\rho}-\frac{l(l+1)}{\rho^{2}}R+\left(\frac{\lambda}{\rho}-\frac{1}{4}\right)R=0
$$

We can find and plot the wave function using odeint, if we knew $\lambda$ and
$dR/d\rho$ at some point. This is an *eigenvalue* problem-
there only exist (finite) solutions for particular values of $\lambda$.
As you probably learned in quantum mechanics, the eigenvalues are
$\lambda=1,2,3,$... 




1. First consider $\lambda=1, l=0.$ The boundary conditions that you want
are that $R=0$ at infinity, and R is finite everywhere. These boundary
conditions are a little tricky to apply numerically- what will end
up happening is that R will diverge due to numerical instabilities
at some point. Start by picking a very large $\rho-$say about 20
and putting in an arbitrary value for R, and $dR/d\rho\sim-R/2$ and
integrating in towards 0. Explore what happens to the solution if
you make $\lambda$ slightly bigger or smaller than one. For each value
of $\lambda$ explore what happens as you vary $dR/d\rho$. You probably
want to use a logarithmic scale for $\rho-$ say going from 20 to 1e-9.
You should be able to find a set of initial conditions/eigenvalues
where the wave function stays relatively constant down to $\rho\sim$ 1e-8.
Look up the wave functions and overlay the analytical result. (Note
that we are free to scale the amplitude of R by a constant, 
so we can always normalize the wave function later. For now, don't
worry about any overall scale factor. )

```python
# You need to run this for interact_manual to work!!

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

from scipy.integrate import odeint

from ipywidgets import interact_manual, fixed

from scipy.special import genlaguerre
from scipy.special import sph_harm
from math import factorial
from scipy.integrate import quad
from scipy.interpolate import interp1d
```

```python
mpl.rc('figure',dpi=250)
mpl.rc('text',usetex=True)
```

```python
'''
Question 1. 

- Numerically solve the schrodinger equation.
- Apply the boundary conditions (rho = 20, dR/d rho = -1/2)
- Vary lambda by a tiny bit.
- For each lambda, vary dR d rho. 

'''

def question_1_integration(lambda_0=1, l=0, dR_0=-1/2):
    '''
    This function solves the above Schrodinger equation and returns the solution.
    '''
    rho_0 = 20

    def main(Y, rho):
        R = Y[0] 
        dRd_rho = Y[1]

        dYd_rho = np.zeros(2)

        dYd_rho[0] = dRd_rho
        dYd_rho[1] = -2/rho*dRd_rho + l*(l+1)/(rho**2)*R - (lambda_0/rho - 1/4)*R
        return dYd_rho
    
    R_0 = 1
    dR_0 = dR_0
    rho = np.logspace(np.log10(20), -9)
    sol = odeint(main, [R_0, dR_0], rho)
    return sol
```

```python
def analytical(rho, theta, phi, n=1, l=0, m=0):
    '''
    This function plots the analytical result. It doesn't seem that this function is normalized.
    
    Update: This function uses units of SI, but I think after normalization it is the same. The question does not ask for 
    normalization, but in the interact we can see that if we turn on my hard coded normalizationfunction
    the general shape of the graph is the same, i.e. the graphs are scaled up by a constant which is the normalization factor
    '''
    reduced_bohr_radius = 5.291e-11
    
    L = genlaguerre(n - l - 1, 2*l + 1)
          
    return np.sqrt((2/(n*reduced_bohr_radius))**3*factorial(n - l - 1)/(2*n*factorial(n+l)))*np.exp(-rho/2)*rho**l*L(rho)*sph_harm(m, l, phi, theta)
```

```python
# Verifying that the numerical solution agrees with the analytical solution. 
# There is a difference because I didn't normalize them.

rho = np.logspace(np.log10(20), -9)
theta = 0
phi = 0
plt.loglog(rho, analytical(rho, theta, phi), label='analytical')
plt.loglog(rho, question_1_integration()[:, 0], label='numerical')
plt.xlabel('log($\\rho$)')
plt.ylabel('R($\\rho$)')
plt.title(f'Analytical Log Log plot of $\\psi(x)$ for $\\lambda = {1}$ , $l = {0}$')
plt.legend()
```

```python
# Verifying that the numerical solution agrees with the ground state hydrogen.
# The next two figures graph the numerical and the analytical solution.

rho = np.logspace(np.log10(20), -9)
plt.plot(rho, question_1_integration()[:, 0], label='numerical')

plt.xlabel('$\\rho$')
plt.ylabel('$\\psi(x)$')
plt.title('$\\psi(x)$ with $\\lambda = 1, l = 0$')
plt.legend()

plt.figure()
plt.plot(rho, analytical(rho, theta, phi), label='analytical')
plt.xlabel('$\\rho$')
plt.ylabel('$\\psi(x)$')
plt.title('$\\psi(x)$ with $\\lambda = 1, l = 0$')
plt.legend()
```

```python
# Varying lambda around l=0, lambda_0=1

def vary_lambda(l=0, lambda_0=1, dR_0 = -1/2, normalize=True, analytic=True):
    '''
    This function plots the numerical solution for varying lambda. 
    
    Normalization might cause the function to fail, but usually it works pretty well.
    '''
    
    lambda_0pm = 0.1
    lambda_0_range = np.linspace(lambda_0-lambda_0pm, lambda_0+lambda_0pm, num = 5)
    
    rho = np.logspace(np.log10(20), -9)
    
    sols = []
    for lambda_0_i in lambda_0_range:
        sol = question_1_integration(lambda_0_i, l, dR_0 = dR_0)
        
        if normalize:
            sol[:, 0] = sol[:,0]/quad(interp1d(rho, sol[:, 0], kind='cubic'), 0, 20)[0]
        sols.append(sol)
        
    plt.figure()
    for sol, lambda_0_i in zip(sols, lambda_0_range):
        plt.loglog(rho, sol[:, 0], label=f'$\lambda$ = {round(lambda_0_i, 3)}')

    if analytic:
        asol = np.real(analytical(rho, theta, phi))
        if normalize:
            asol = asol/quad(interp1d(rho, asol, kind='cubic'), 0, 20)[0]

        plt.loglog(rho, asol, label='analytical')
    
    plt.xlabel('log($\\rho$)')
    plt.ylabel('R($\\rho$)')
    plt.title(f'Log Log plot of $\\psi(x)$ for $\\lambda = {round(lambda_0, 3)}$ , $l = {l}$, $dR/d\\rho = {dR_0}$')
    plt.legend()
```

```python
#You will need to import interact_manual from ipywidgets, i.e. run the preamble
#This function graphs l=0, lambda_0 = 1

interact_manual(vary_lambda, l=fixed(0), lambda_0=fixed(1), dR_0=fixed(-1/2), normalize=True, analytic=True)
```

```python
# Varying dRdrho around l=0,lambda_0=1

# For a given lambda, vary dRdrho

def vary_dRdrho(l=0, lambda_0=1, dR_0 = -1/2, normalize=False, analytic=True):
    '''
    This function plots the numerical solution for varying dRdrho. 
    
    Normalization is very unstable and sometimes causes this function to fail.
    '''
    
    dR_0pm = 0.05
    dR_0_range = np.linspace(dR_0 - dR_0pm, dR_0 + dR_0pm, num = 5)
    
    rho = np.logspace(np.log10(20), -9)
    
    sols = []
    for dR_0_i in dR_0_range:
        sol = question_1_integration(lambda_0, l, dR_0 = dR_0_i)
        
        if normalize:
            sol[:, 0] = sol[:, 0]/quad(interp1d(rho, sol[:, 0], kind='cubic'), 0, 20)[0]
        sols.append(sol)
        
    plt.figure()
    for sol, dR_0_i in zip(sols, dR_0_range):
        plt.loglog(rho, sol[:, 0], label=f'$dR/d\\rho$ = {round(dR_0_i, 3)}')

    if analytic:
        asol = np.real(analytical(rho, theta, phi))
        if normalize:
            asol = asol/quad(interp1d(rho, asol, kind='cubic'), 0, 20)[0]

        plt.loglog(rho, asol, label='analytical')

    plt.xlabel('log($\\rho$)')
    plt.ylabel('R($\\rho$)')
    plt.title(f'Log Log plot of $\\psi(x)$ for $\\lambda = {round(lambda_0, 3)}$ , $l = {l}$')
    plt.legend()
```

3. The physical potential isn't $1/r$ inside the nucleus. Instead, we
should consider the nucleus as a sphere of uniform charge density-
so the potential is 0 at $r=0$, and is proportional to $r$ until we get
to the outside of the nucleus, at $r=1.2\times 10^{-15}$ m. Change the potential
and compare the new eigenvalues and wavefunctions to the old ones.
For this case there are no analytic solutions.

$$
\frac{d^{2}R}{d\rho^{2}}+\frac{2}{\rho}\frac{dR}{d\rho}-\frac{l(l+1)}{\rho^{2}}R+\left(\frac{\lambda}{\rho}-\frac{1}{4}\right)R=0
$$

$$
\begin{array}{cc}
\rho=\sqrt{\frac{-8\mu E}{\hbar^{2}}}r & \lambda=\frac{Ze^{2}}{\hbar}\sqrt{\frac{-\mu}{2E}}\end{array};
$$ 

```python
r0 = 1.2e-15
E = -13.6*1.60217662e-19
h = 1.0545718e-34
mu = 9.10938356e-31
rho0 = np.sqrt(-8*mu*E/h**2)*r0

def potential(lambda_0, rho):
    if rho < rho0:
        return lambda_0*rho
    elif rho >= rho0:
        return lambda_0/rho
```

```python
rho = np.linspace(0, 20)
potential = np.vectorize(potential)

plt.plot(rho, potential(1, rho))
plt.xlabel('$\\rho$')
plt.ylabel('$V(\\rho)$')
plt.title('$V(\\rho)$ as a function of $\\rho$')
```

```python
def question_3_integration(lambda_0=1, l=0, dR_0=-1/2):
    '''
    This function solves the above Schrodinger equation and returns the solution.
    '''
    rho_0 = 20

    def main(Y, rho):
        R = Y[0] 
        dRd_rho = Y[1]

        dYd_rho = np.zeros(2)

        dYd_rho[0] = dRd_rho
        dYd_rho[1] = -2/rho*dRd_rho + l*(l+1)/(rho**2)*R - (potential(lambda_0, rho) - 1/4)*R
        return dYd_rho
    
    R_0 = 1
    dR_0 = dR_0
    rho = np.logspace(np.log10(20), -9)
    sol = odeint(main, [R_0, dR_0], rho)
    return sol
```

```python
def question_3_compare(lambda_0 = 1, l=0, dR_0 = -0.5):
    rho = np.logspace(np.log10(20), -9)

    sol = question_3_integration(lambda_0 = lambda_0, l=l, dR_0 = dR_0)
    plt.loglog(rho, sol[:, 0], label='with Nucleus correction', alpha=0.7)

    sol = question_1_integration(lambda_0 = lambda_0, l=l, dR_0 = dR_0)
    plt.loglog(rho, sol[:, 0], label='without Nucleus correction', alpha=-0.7)

    plt.xlabel('$\\rho$')
    plt.ylabel('$\\psi(\\rho)$')
    plt.title('Comparison of numerical solutions of $\\psi(\\rho)$')
    plt.legend()
    
question_3_compare()
```
