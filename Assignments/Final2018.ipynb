The actual value of $I_0$ doesn't matter here since we take a ratio- so you just need to throw a large number of muons with this angular distribution.  All three detectors are 0.25 m x 1 m.  

```python
def generate_samples(func, bounds, N): 
    """
    Assignment 5: function for sampling from a custom 1d distribution.
    
    Input:  func - any function
            bounds - tuple (a, b)
    
    Output: 1d samples of size N
    """
    a, b = bounds
    x = np.linspace(a, b)
    
    f_max = max(func(x))
    
    x_rand = np.random.uniform(a, b, size=N)
    y_rand = np.random.uniform(0, f_max, size=N)
        
    samples = x_rand[y_rand < func(x_rand)]

    return samples.reshape(len(samples), 1)

func = lambda x: np.cos(x)**2
theta = generate_samples(func, (0, np.pi/2), N)
plt.hist(theta , bins=50)
theta
```

A pulse of sound is emitted from a position (0,0), and travels with velocity $v=2000 \frac{\rm{m}}{\rm{s}}$.  The sound is reflected from a layer, which described by \begin{equation}(x,z)=(x,f(x))=(x,ax^3+bx^2+cx+d).\end{equation}  Here x is the horizontal position and z is the depth into the earth. 

The time it takes the sound to move from (0,0) on the surface to a point on the layer $\vec{p}$ is:\begin{equation}
t_1=\frac{|\vec{p}|}{v}\end{equation}

To do the reflection, we set up coordinates with x along the boundary at point $\vec{p}=(x,f(x))$ 

\begin{align}
\hat{i}=\frac{(1,df/dx)}{|(1,df/dx)|}\\
\hat{k}=-\frac{\vec{p}-({\vec{p}\cdot\hat{i}})\hat{i}}{|\vec{p}-({\vec{p}\cdot\hat{i}})\hat{i}|}
\end{align}

In this coordinate system the reflected wave just continues in the $\hat{i}$ direction, but reflects in the $\hat{k}$ direction.   So the final direction  \begin{equation}\hat{n}=\frac{(\vec{p}\cdot\hat{i})\hat{i}-(\vec{p}\cdot\hat{k})\hat{k}}{|(\vec{p}\cdot\hat{i})\hat{i}-(\vec{p}\cdot\hat{k})\hat{k}|}.\end{equation}  

To find the position $\vec{x_s}$ and time $t$ where the pulse returns to the surface we calculate
\begin{align} t_2=-\frac{f(x)}{v\cdot n_z}\\
\vec{x_s}=(x_s, z_s)=p+t_2v\hat{n}\\
t=t_1+t_2
\end{align}

A baseball, with diameter 7.4 cm and mass 146 g, has been hit a distance of about 150 meters.  After the ball leaves the bat, the only forces are those due to gravity and air resistance (we ignore the possibility of lift forces due to the spin of the ball).  Assuming that there is no wind, and the density of the air is $\rho=1.225 \frac{\rm{kg}}/{\rm{m}^3}$, and the magnitude of the turbulent drag, $F_t$, is
\begin{equation}
F_t=\frac{1}{2}\rho v^2AC_D
\end{equation}
where $A=\pi r^2$ is the cross sectional area of the sphere, $C_D$ is the drag coefficient.

The drag force is in the opposite direction to the velocity.

```python
airDensity=1.225 # kg/m**3
airPressure=101.325 #kPa
radius=0.037
airViscosity=1.827e-5   #Pa-s  

RN=np.array([1,4,10,40,100,400,1000,2000,4000,10000,40000,100000,280000,400000,500000])
DC=np.array([24,8,4,1.7,1.1,0.6,0.45,0.4,0.35,0.4,0.45,0.42,0.39,0.08,0.08])

mass = 0.146
g=9.81
```

```python
from scipy.interpolate import interp1d

reynolds_to_drag = interp1d(RN, DC)
```

```python
def TurbulentDrag(velocity):
    reynolds = 2*airDensity*np.linalg.norm(velocity)*radius/airViscosity
    dragCoef = reynolds_to_drag(reynolds)
    Fmag = 0.5*airDensity*np.dot(velocity, velocity)*np.pi*radius**2*dragCoef
    
    return -Fmag*velocity/np.linalg.norm(velocity)
```

```python
def Gravity():
    return mass*g*np.array([0, 0, -1])
```

```python
y = np.arange(10)

def ball(y, t):
    position = y[0:3]
    velocity = y[3:6]
    
    dydt = np.empty(6)
    
    dydt[0:3] = velocity
    dydt[3:6] = TurbulentDrag(velocity) + Gravity() 
    
    return dydt 


```

```python
initial_v = 100
theta = np.pi/4

y0 = np.array([0, 0, 0, initial_v*np.cos(theta), 0, initial_v*np.sin(theta)])
t = np.linspace(0, 100, 1000)
y = odeint(ball, y0, t)
```

## Statistics and Monte Carlo

A beam of particles is incident on a target and produces radioactive atoms.  In typical operation it produces R atoms/second.  The way in which the atoms are produced is *independent*- that is, one atom is produced at a time and there is no connection between the rate at which atoms are produced doesn't depend on how many atoms are there already.  

This problem involves writing various Monte Carlo calculations that simulate the statistics of this situation.  

A simple way to approximate this is to break the time interval into very small time steps, so that in each step only $\lambda<<1$ atoms are produced.   We throw a random number for each time step, and generate an atom if that number is less than $\lambda$.

Consider the case when R=10 atoms/second.  Pick $\lambda=0.05,$ so the time interval will be $\delta=\lambda/R=0.005 \text{s}.$  Calculate (by MC) the number of atoms created in a second.  Run the 10000 experiments, and plot the distribution of the number of atoms created in a second interval.



```python
lambda_0 = 0.05
R = 10
delta = lambda_0/R
delta

n_throws = int(1/delta)
print(n_throws)

n_experiments = 10000

# counts in a second per experiment

n_counts = []
for i in range(n_experiments):
    n_in = 0
    for j in range(n_throws):
        if np.random.rand() < lambda_0:
            n_in += 1
    n_counts.append(n_in)
```

To do the problem exactly, we generate an expontial time separation between adjacent events.  That is\begin{equation}P(t+dt)=P(t)(1-dt*R)\rightarrow\frac{dP}{dt}=-RP\rightarrow P=e^{-Rt}\end{equation}

Thus, the probability distribution for time intervals between events is $P=e^{-Rt}.$  Generate 10000 1 second experiments with this distribution, and plot the number of events in a second.  Also plot the number of events in a 100,000 experiments with a 0.005 second interval.



```python
R = 10

func = lambda t: np.exp(-R*t)

n_experiments = 10000

times = []
n_events_per_second = []
for i in range(n_experiments):
    running_time = 0
    n_events = 0
    while running_time < 1:
        running_time += generate_samples(func, (0, 1), 300)[0]
        n_events += 1
    times.append(running_time)
    n_events_per_second.append(n_events)
    
```
